9. Plot different features against one another

```python
white_df.quality.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title("Distribution of Wine Quality (White Wine)")
plt.ylabel('Number of Wines')
plt.xlabel('Quality');
```

```python
plt.figure(figsize=(10,6))
sns.histplot(white_df['alcohol'])
plt.title("Distribution of Alcohol Content (White Wine)")
plt.xlabel("Alcohol")
plt.ylabel("Frequency")
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x=white_df['alcohol'], y=white_df['quality'])
plt.title("Relationship between Alcohol and Quality (White Wine)")
plt.xlabel("Alcohol")
plt.ylabel("Quality")
```
